```python
def wunderLink(url):
    # Commands related to the webdriver (not sure what they do, but I can guess):
    #bi = chromedriver_binary(r'C:\Program Files (x86)\Google\Chrome\Application\chrome.exe')
    options = Options()
    options.binary_location = 'C:\Program Files (x86)\Google\Chrome\Application\chrome.exe'
    br = webdriver.Chrome(options = options, executable_path="C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe")

    # This starts an instance of Firefox at the specified URL:
    br.get(url)

    # I understand that at this point the data is in html format and can be
    # extracted with BeautifulSoup:
    sopa = BeautifulSoup(br.page_source, 'lxml')

    solar = sopa.find('lib-history-table')
    
    br.quit()
    
    wunderExtract = extractWunder(solar)
    
    return wunderExtract
```

```python
def extractWunder(htmlData):
    columnNames = ['time', 'temperature', 'humidity', 'windDirection', 'speed', 'gust', 'pressure', 'precipitationRate', 'precipitationAccumulation', 'solarIntensity']
    wunderData = pd.DataFrame(columns = columnNames)
    for div_tag_root in htmlData:                                 # Navigating through the long row of HMTL-tags to obtain
        for div_tag_table_wrapper in div_tag_root:                # the intended values.
            for div_tag_scrollable in div_tag_table_wrapper:
                for table_tag in div_tag_scrollable:
                    for tbody_tag in table_tag:
                        if len(tbody_tag) > 2:
                            for tr_tag in tbody_tag:
                                if hasattr(tr_tag, 'contents'):
                                    x = 0
                                    cache = []
                                    while x < len(tr_tag.contents):
                                        if x in [0, 4, 11]:
                                            cache.append(str(tr_tag.contents[x]).split('>')[2].split('<')[0])
                                        if x in [1, 3, 5, 6, 7, 8, 9]:
                                            cache.append(str(tr_tag.contents[x]).split('>')[7].split('<')[0])
                                        x += 1
                                    cache = pd.DataFrame([cache], columns = columnNames, index = None)
                                    wunderData = pd.concat([wunderData, cache])
    return wunderData
```

```python
def convertUnits(df):
    for column in df:
        if column in ['time, windDirection']:
            continue
        else:
            df[column] = pd.to_numeric(df[column], 'ignore', 'float')
            
    df['pressure'] = 33.865 * df['pressure']
    df['temperature'] = round(((df['temperature'] - 32) * (5/9)), 2)
    df['speed'] = round((df['speed'] * 0.44704), 2)
    df['gust'] = round((df['gust'] * 0.44704), 2)
    df['precipitationRate'] = round((df['precipitationRate'] * 25.4), 2)
    df['precipitationAccumulation'] = round((df['precipitationAccumulation'] * 25.4), 2)
    
    return df
```
